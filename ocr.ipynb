```python
import cv2
!pip install pytesseract
import pytesseract
from PIL import Image
from matplotlib import pyplot as plt
```

```python
def display(im_path):
    dpi = 80
    im_data = plt.imread(im_path)

    height, width  = im_data.shape[:2]

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()
```

```python
img = cv2.imread("/content/im.png")
```

```python
display("/content/im.png")
```

```python
inverted_image = cv2.bitwise_not(img)
cv2.imwrite("/content/inverted.jpg",inverted_image)
```

```python
display("/content/inverted.jpg")
```

```python
def grayscale(image):
  return cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
gray_image = grayscale(img)
cv2.imwrite('/content/gray.jpg',gray_image)
```

```python
display('/content/gray.jpg')
```

```python
thresh, bw_image = cv2.threshold(gray_image, 150,150,cv2.THRESH_BINARY)
cv2.imwrite("/content/bw_image.jpg",bw_image)
```

```python
display("/content/bw_image.jpg")
```

```python
def noise_removal(image):
    import numpy as np
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    image = cv2.medianBlur(image, 3)
    return (image)
```

```python
no_noise = noise_removal(bw_image)
cv2.imwrite("/content/no_noise.jpg",no_noise)
```

```python
display("/content/no_noise.jpg")
```

```python
def thin_font(image):
    import numpy as np
    image = cv2.bitwise_not(image)
    kernel = np.ones((2,2),np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    image = cv2.bitwise_not(image)
    return (image)
```

```python
thinfont_image = thin_font(no_noise)
cv2.imwrite("/content/eroded_image.jpg",thinfont_image)
```

```python
display("/content/eroded_image.jpg")
```

```python
def thick_font(image):
    import numpy as np
    image = cv2.bitwise_not(image)
    kernel = np.ones((2,2),np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    image = cv2.bitwise_not(image)
    return (image)
```

```python
dilated_image = thick_font(no_noise)
cv2.imwrite("/content/dilated_image.jpg", dilated_image)
```

```python
display("/content/dilated_image.jpg")
```
